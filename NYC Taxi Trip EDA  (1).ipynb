# UVA Analysis#

```python
def UVA_analysis (data, var_group):
    size = len(var_group)
    plt.figure(figsize = (18,6),dpi = 100)
    
    for j,i in enumerate (var_group):
        
        if pd.api.types.is_numeric_dtype(data[i]):
    
            maxi = round(data[i].max(),2)
            mini =  round(data[i].min(),2)
            ran = round( data[i].max()-data[i].mean(), 2)
            mean =  round(data[i].mean(),2)
            median =  round( data[i].median(),2)
            st_dev =  round(data[i].std(),2)
            skew =  round(data[i].skew(),2)
            kurt  =  round(data[i].kurtosis(),2)
        elif pd.api.types.is_timedelta64_dtype(data[i]):
            data_in_seconds = data[i].dt.total_seconds()
            maxi = (data[i].max())
            mini =  (data[i].min())
            ran = (data[i].max()-data[i].mean())
            mean =  (data[i].mean())
            median =  ( data[i].median())
            st_dev =  (data[i].std())
            skew =  None
            kurt  =  None
        else:
            continue 
        
        points =round(mean-st_dev,2),round(mean+st_dev,2)
        
        plt.subplot(1,3,j+1)
        sns.kdeplot(x = data[i] , fill = True)
        sns.lineplot (x = points, y = [0,0], color = 'blue', label = f'with_in_one_std')
        sns.scatterplot(x = [maxi,mini], y =[0,0], label = f'max:{maxi},min:{mini}')
        sns.scatterplot(x = [mean], y = [0], label = f'mean:{mean}')
        sns.scatterplot(x = [median] , y = [0], label = f'median:{median}')
        plt.xlabel(f'{i}',fontsize = 20 )
        plt.ylabel ('density')
        plt.title(f'with_in_one_stdev= {points[0]} to {points[1]}; skew={skew};\n kurtosis= {kurt}; range = {ran}')
```

```python
def Outliers (data, var_group, outliers = True):
    size = len(var_group)
    plt.figure(figsize = (7*size,4),dpi = 100)
    
    for j,i in enumerate (var_group):
        Q1 = round(data[i].quantile(0.25),2)
        Q3 = round(data[i].quantile(0.75),2)
        IQR = (Q3-Q1)
        median = round(data[i].median(), 2)
        upper_bound = Q3+(1.5*IQR)
        lower_bound = Q1-(1.5*IQR)
        
        outlier_high = len(data[i][data[i]>upper_bound])
        outlier_low = len(data[i][data[i]<lower_bound])
        
        
        if outliers ==True:
        
            plt.subplot (1,size,j+1)

            sns.boxplot(x =data[i],orient = 'v' )
            plt.ylabel(f'{[i]}')
            plt.title(f'with outliers\n IQR = {IQR}, Median = {median}\n 1st_and_3rd_Quartile = {Q1,Q3},\n low and high outliers = {outlier_high,outlier_low}')
            
        else:
            data_2 = data[var_group] 
            data_2.loc[data_2[i]>Q3,i]= Q3+1
            data_2.loc[data_2[i]<Q1,i]= Q1-1
            
            plt.subplot (1,size,j+1)

            sns.boxplot(x =data_2[i],orient = 'v' )
            plt.ylabel(f'{[i]}')
            plt.title(f'with outliers\n IQR = {IQR}, Median = {median}\n 1st_and_3rd_Quartile = {Q1,Q3}')
            
```

```python
from numpy import sqrt,abs ,round
from scipy.stats import norm
```

```python
def Twosamp_Z(x1,x2,sig1,sig2,n1,n2):
    over_sigma = sqrt(sig1**2/n1)+sqrt(sig2**2/n2)
    z = (x1-x2)/over_sigma
    pval = 2*(1-norm.cdf(abs(z)))
    return pval
```

```python
def Twosamp_T(X1,X2,sd1,sd2,n1,n2):
    from scipy.stats import t as t_dist
    over_sigma = sqrt(sd1**2/n1)+ sqrt(sd2**2/n2)
    t = (X1-X2)/over_sigma
    df = n1+n2-2
    tval = 2*(1-t_dist.cdf(abs(t),df))
    return tval
```

```python
def bivariate_cont_cat(data,cont,cat,category):
    x1 = data[cont][data[cat]==category]
    x2 = data[cont][data[cat]!=category]
    
    n1,n2 = x1.shape[0],x2.shape[0]
    if n1 == 0 or n2 == 0:
        print("Error: One of the groups has zero observations.")
        return
    
    sd1,sd2 = x1.std(), x2.std()
    m1,m2 = x1.mean(),x2.mean()
    
    z_p_val = Twosamp_Z(m1,m2,sd1,sd2,n1,n2)
    t_p_val = Twosamp_T(m1,m2,sd1,sd2,n1,n2)
    
    table = pd.pivot_table(data = data , values = cont,columns = cat, aggfunc = np.mean)
    
    plt.figure(figsize = (15,6), dpi = 140)
    plt.subplot(1,2,1)
    
    sns.barplot(x= cat,y = cont, data = data)
    plt.xlabel(cat)
    plt.title ( f'z_test_p value = {z_p_val}\n T_test_p_value={t_p_val} \n{table}')
    
    plt.subplot(1,2,2)
    sns.boxplot(x= cat , y= cont, data = data )
    plt.title ('categorical_boxplot')
```
